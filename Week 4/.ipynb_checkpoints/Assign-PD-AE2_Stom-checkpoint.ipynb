```python
# create a 3D weighting/projection matrix (m is number of dimensions)
def FETrans(mF, m=1):
    size = np.shape(mF)[0]
    W = np.identity(size) - mF @ np.linalg.inv(mF.T@ mF/size)@ mF.T/size
    if m > 1:
        W = np.broadcast_to(W, (m,) + W.shape)
    return W

# prepare X and Y data
def PrepareData(df, iSizeN, iSizeT, sCol):
    mDataY  = np.transpose(np.asmatrix(df[sCol[0]].values).reshape(iSizeN,iSizeT))
    iSizeK = np.int(np.size(sCol)-1)
    aX = np.empty((iSizeK,iSizeT,iSizeN))
    for k in range (0,iSizeK):
        aX[k,:,:] = np.transpose(np.asmatrix(df[sCol[k+1]].values).reshape(iSizeN,iSizeT))
    return (mDataY,aX)
```

```python
N = len(np.unique(df.ID))
T = len(np.unique(df.Time))
sCol = ['DeltaS', 'DeltaS_lag', 'DeltaY_lag','DeltaY','DeltaY_lead']
K = len(sCol)-1

Y, X = PrepareData(df, iSizeN=N, iSizeT=T, sCol=sCol)

# Y and X fixed for time effect. Can be used in the FE, HPJ and MG functions instead of X,Y as well.
Y_dot = Y @ FETrans(np.ones([N,1]), 1)
X_dot = X @ FETrans(np.ones([N,1]), K)
```

```python
# return fixed effect estimator and if CCM=True also returns â€œclustered covariance matrix 
def FE(X, Y, CCM=False):
    _, T, N = X.shape
    W_T = FETrans(np.ones([T,1]), N)
    
    # slide 1.17 eq. (17)
    XWX = (X.transpose((2,0,1)) @ W_T @ X.transpose((2,1,0))).sum(axis=0)
    XWY = (X.transpose((2,0,1)) @ W_T @ np.expand_dims(Y.T, axis=2)).sum(axis=0)
    beta = (np.linalg.inv(XWX) @ XWY  )[:,0]

    if CCM:
        # slide 1.44 eq. (22)
        res = (Y - (X.transpose((1,2,0)) @ beta))
        XWe = ((X.transpose(2,0,1) @ W_T) @ np.expand_dims(res.T, axis=2))[:,:,0]   
        ccm = XWe.T @ XWe  / N
        return beta, ccm
    
    return beta 

# returns Half Panel Jackknife estimator
def HPJ(X, Y):
    beta = FE(X, Y)
    X1, X2 = np.hsplit(X, 2)
    Y1, Y2 = np.vsplit(Y, 2)
    beta1, beta2 = (FE(x, y) for x, y in [(X1, Y1), (X2, Y2)])

    return 2*beta - 0.5*(beta1 + beta2)

# returns mean group estimator using either the FE or HPJ methods
def MG(X, Y, method):
    _, _, N = X.shape
    beta = np.zeros([K, N])
    
    # slide 1.31 eq. (18)
    for i in range(N):
        beta[:,i] += method(np.expand_dims(X[:,:,i], axis=2), Y[:,i])
    
    return beta.mean(axis=1), beta

# calculate estimates
beta_FE = FE(X, Y)
beta_2WFE = FE(X_dot, Y_dot)

beta_FE_HPJ = HPJ(X, Y)
beta_2WFE_HPJ = HPJ(X_dot, Y_dot)

beta_MG, _ = MG(X, Y, FE)
beta_HPJ_MG, _ = MG(X, Y, HPJ)
```

```python
# returns 2Ã—K array of 0.025% and 0.975% quantiles of bootstrapped beta estimates
# mg=1 means that the mean estimator is used (in combination with the method is FE or HPJ)
def bootstrap_CI(X, Y, B, method, mg=False):
    np.random.seed(1)
    K, _, N = X.shape
    if mg:    
        _, beta = MG(X, Y, method)
        
    betaB = np.zeros([B, K])
    index_B = np.random.permutation(np.repeat(np.arange(N),B)).reshape((B,N))
    for b in range(B):
        index = index_B[b,:]
        if mg:
            betaB[b,:] = beta[:,index].mean(axis=1)
        else:
            XB = X[:,:,index]
            YB = Y[:,index]
            betaB[b,:] = method(XB, YB)
        
    return np.quantile(betaB, [0.025, 0.975], axis=0)

# returns a 2Ã—K array of approximate normal 95% upper and lower bounds of FE estimator
def normal_CI(X, Y):
    _, T, N = X.shape
    beta, ccm = FE(X, Y, 1)
    W_T = FETrans(np.ones([T,1]), N)
    
    # slide 1.44 eq. (21)
    W = (X.transpose((2,0,1)) @ W_T @ X.transpose((2,1,0))).sum(axis=0) / (N*T)
    W1 = np.linalg.inv(W)
    offset = np.sqrt(np.diagonal(W1 @ ccm @ W1)) / np.sqrt(N) * 1.96
    return np.array([-offset.T, offset.T])

# returns a 2Ã—K array of approximate normal 95% upper and lower bounds of FE-MG estimator
def normal_MG_CI(X, Y, method):
    _, _, N = X.shape
    betaN, beta = MG(X, Y, method)
    
    betaN = np.repeat(np.array([betaN]), N, 0).T # matrix with N columns of betaN
    sigma = (beta - betaN) @ (beta - betaN).T / (N-1)
    
    offset = np.sqrt(np.diagonal(sigma)) / np.sqrt(N) * 1.96
    return np.array([-offset.T, offset.T])
    
# calculate and format confidence intervals
methods = ['FE bs','2WFE bs','FE-HPJ bs','2WFE-HPJ bs','FE normal','2WFE normal']
methods_MG = ['FE_MG normal','FE-MG bs','HPJ-MG bs']
index = [(a, b) for a in sCol[1:] for b in ['lower','upper']]
index = pd.MultiIndex.from_tuples(index)

CI = bootstrap_CI(X, Y, 400, FE).T.flatten()
CI = np.vstack((CI, bootstrap_CI(X_dot, Y_dot, 400, FE).T.flatten()))

CI = np.vstack((CI, bootstrap_CI(X, Y, 400, HPJ).T.flatten()))
CI = np.vstack((CI, bootstrap_CI(X_dot, Y_dot, 400, HPJ).T.flatten()))

CI = np.vstack((CI, normal_CI(X, Y).T.flatten()))
CI = np.vstack((CI, normal_CI(X_dot, Y_dot).T.flatten()))

CI_MG = normal_MG_CI(X, Y, FE).T.flatten()
CI_MG = np.vstack((CI_MG, bootstrap_CI(X, Y, 400, FE, mg=True).T.flatten()))
CI_MG = np.vstack((CI_MG, bootstrap_CI(X, Y, 400, HPJ, mg=True).T.flatten()))

CI = pd.DataFrame(data=CI.T, index=index, columns=methods)
CI_MG = pd.DataFrame(data=CI_MG.T, index=index, columns=methods_MG)
```
